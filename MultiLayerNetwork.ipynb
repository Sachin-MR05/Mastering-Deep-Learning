```python
n =2
num_hidden_layer =2
m = [2,2]
num_node_output =1
```

```python
import numpy as np
num_nodes_previous =n
```

```python
network ={}
```

```python
def initialize_network(n,num_hidden_layer,m,num_node_output):
  network ={}
  num_nodes_previous =n
  for layer in range(num_hidden_layer+1):
    if layer ==num_hidden_layer:
      layer_name = 'output_layer'
      num_nodes = num_node_output
    else:
      layer_name = 'layer_{}'.format(layer+1)
      num_nodes = m[layer]

    network[layer_name]={}
    for node in range(num_nodes):
      node_name = 'node_{}'.format(node+1)
      network[layer_name][node_name]={
        'weight':np.around(np.random.uniform(size= num_nodes_previous),decimals=2),
        'bias':np.around(np.random.uniform(size = 1),decimals =2),
    }
      num_node_previous = num_nodes
    return network
```

```python
initialize_network(n,num_hidden_layer,m,num_node_output)
```

```python
def compute_weighted_sum(input,weights,bias):
  return np.sum(input*weights)+bias
```

```python
from random import seed
import numpy as np
```

```python
np.random.seed(12)
inputs = np.around(np.random.uniform(size=2),decimals=2)
print(inputs)
```

```python
def node_activatio(weighted_sum):
  return 1.0/(1.0+ np.exp(-1 * weighted_sum))
```

```python
def forward_propagation(network,inputs):
  layer_inputs = list(inputs)
  for layer in network:
    layer_data = network[layer]
    layer_outputs =[]
    for layer_node in layer_data:
      node_data = layer_data[layer_node]
      node_output = node_activatio(compute_weighted_sum(layer_inputs,node_data['weight'],node_data['bias']))
      layer_outputs.append(np.around(node_output[0],decimals=4))
    if layer != "output":
      print('The outputs of the nodes in hidden layes number {} is {}'.format(layer.split('_')[1],layer_outputs))

    layer_inputs = layer_outputs
  network_predictions = layer_outputs
  return network_predictions
```

```python
predictions = forward_propagation(network,inputs)
predictions
```
