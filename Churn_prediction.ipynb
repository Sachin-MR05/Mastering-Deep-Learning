```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Telco_Churn.csv")
df = pd.DataFrame(data)
df
```

```python
df = df.drop('customerID', axis=1)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df1 = df.dropna(subset=['TotalCharges'])
df1
```

```python
df1.dtypes
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python

binary_map = {'Yes': 1, 'No': 0, 'No phone service': 0, 'No internet service': 0}
gender_map = {'Male': 1, 'Female': 0}


yes_no_columns = [
    'Partner',
    'Dependents',
    'PhoneService',
    'MultipleLines',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies',
    'PaperlessBilling',
    'Churn'
]


for col in yes_no_columns:
    if col in df1.columns:
        df1[col] = df1[col].map(binary_map)

# Apply the gender map
if 'gender' in df1.columns:
    df1['gender'] = df1['gender'].map(gender_map)


display(df1.head())
```

```python
display(df1.dtypes)
```

```python
df1 = pd.get_dummies(df1,columns=['InternetService','Contract','PaymentMethod'])
```

```python
import sklearn
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
scaling_colums = ['tenure', 'MonthlyCharges', 'TotalCharges']
scaler = MinMaxScaler()
df1[scaling_colums] = scaler.fit_transform(df1[scaling_colums])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df1.drop('Churn', axis=1)
y = df1['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split( x,y,test_size=0.2,random_state = 42)
```

```python
for i in X_train:
  print(X_train[i].unique())
```

```python
std_scaler = StandardScaler()
X_train_scaled = std_scaler.fit_transform(X_train)
X_test_scaled = std_scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
```

```python
model = Sequential([
    keras.layers.Dense(128, input_shape=(X_train_scaled.shape[1],), activation='relu'),
    keras.layers.Dense(28, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid') # Changed to sigmoid for binary classification
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train, epochs=5)
```

```python
y_predict =model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
predict = []
for i in y_predict:
  if i <= 0.5:
    predict.append(0)
  else:
    predict.append(1)
```

```python
print(classification_report(y_test,predict))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=predict)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
