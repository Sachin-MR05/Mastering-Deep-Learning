```python
import numpy as np
```

```python
def sigmoid(x):
  return x*(1-x)
```

```python
def sigmoid_derivative(x):
  return x*(1-x)
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0], [1], [1], [0]])
X, y
```

```python
np.random.seed(42)
```

```python
input_size =2
hidden_size =3
output_size=1
```

```python
w1 = np.random.uniform(size=(input_size,hidden_size))
b1 = np.zeros((1,hidden_size))
b1
```

```python
w2 = np.random.uniform(size=(hidden_size,output_size))
b2 = np.zeros((1,output_size))
w2
```

```python
learning_rate =0.1
```

```python
import numpy as np
for epoch in range(100):
  Z1 = np.dot(X,w1)+b1
  A1 = sigmoid(Z1)

  Z2 = np.dot(A1,w2)+b2
  A2 = sigmoid(Z2)

  error = y-A2
  loss = np.mean(np.square(error))

  d_A2 = error * sigmoid(A2)

  error_hidden = d_A2.dot(w2.T)
  d_A1 = error_hidden * sigmoid(A1)

  w2 += A1.T.dot(d_A2)*learning_rate
  b2 += np.sum(d_A2,axis=0,keepdims =True)*learning_rate
  w1 += X.T.dot(d_A1)*learning_rate
  b1 += np.sum(d_A1, axis=0,keepdims=True)*learning_rate

  if epoch%10==0:
    print(f"Epoch {epoch},Loss: {loss:.4f}")
```
